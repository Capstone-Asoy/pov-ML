```python
# Mengonversi kolom 'tahun_terbit' dan 'jml_halaman' menjadi integer
df_indo['tahun_terbit'] = df_indo['tahun_terbit'].astype(int)
df_indo['jml_halaman'] = df_indo['jml_halaman'].astype(int)
```

```python
# Mencari dan menampilkan baris dengan judul yang berulang
duplicated_titles = df_indo[df_indo.duplicated(subset=['judul'], keep=False)]
# Mengurutkan DataFrame berdasarkan kolom 'judul'
duplicated_titles = duplicated_titles.sort_values(by='judul')
duplicated_titles
```

```python
# Identify rows where the genre column is empty or contains an empty list
empty_genre_rows = df_indo[df_indo['genre'].apply(lambda x: len(eval(x)) == 0 if isinstance(x, str) else True)]

total_rows = len(df_indo)
empty_genre_count = len(empty_genre_rows)
percentage_empty_genre = (empty_genre_count / total_rows) * 100

print(f"There are {empty_genre_count} book titles with an empty genre out of a total of {total_rows} book titles in the dataset.")
print(f"The percentage of book titles with an empty genre is {percentage_empty_genre:.2f}%.")
empty_genre_rows
```

```python
# Fungsi untuk memisahkan genre yang berbentuk daftar string menjadi nilai-nilai individu
from collections import Counter
import ast
def split_genres(genre_str):
    try:
        # Mengubah string list menjadi list python
        genres = ast.literal_eval(genre_str)
        if isinstance(genres, list):
            return genres
        else:
            return [genre_str]
    except:
        return [genre_str]

# Fungsi untuk menghitung genre unik dan total buku
def count_unique_genres(df, col_name='genre_list'):
    # Menggabungkan semua list genre menjadi satu list besar
    all_genres = [genre for sublist in df[col_name] for genre in sublist]

    # Menghitung frekuensi setiap genre
    genre_counts = Counter(all_genres)

    return genre_counts
```

```python
# Terapkan fungsi ke kolom 'genre'
df_indo['genre_list'] = df_indo['genre'].apply(split_genres)

# Hitung genre unik dan total buku
genre_counts = count_unique_genres(df_indo)

# Konversi hasil ke DataFrame
genre_counts_df = pd.DataFrame(genre_counts.items(), columns=['genre', 'jumlah_buku'])
genre_counts_df
genre_counts_df.to_csv('genre_counts.csv', index=False)
```
