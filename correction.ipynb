```python
import requests
from bs4 import BeautifulSoup
```

```python
nocover = "https://dryofg8nmyqjw.cloudfront.net/images/no-cover.png"
nocovering = "https://s.gr-assets.com/assets/nophoto/book/111x148-bcc042a9c91a29c1d680899eff700a03.png"
mynocover = "https://storage.googleapis.com/profileuser/book_image.png"
```

```python
def get_image_src(url):
    try:
        response = requests.get(url)
        response.raise_for_status() 
        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')
        img_element = soup.find('img', {'class': 'ResponsiveImage', 'role': 'presentation'})
        if img_element:
            src = img_element["src"]
            print(f"getted {src}")
            if src == nocover:
                return {'url' : url,'image': mynocover}
            return {'url' : url,'image': src}
        else:
            print("get_nothing")
            return "Image element not found"
    except requests.exceptions.RequestException as e:
        return f"An error occurred: {e}"
    
link = "https://www.goodreads.com/book/show/25995785-d-gray-man-vol-11"
print(get_image_src(link))
```

```python
filtered = final_data[final_data['image'] == nocovering]
```

```python
filtered.info()
```

```python
from concurrent.futures import ThreadPoolExecutor, as_completed
```

```python
max_threads = 20
results = []
```

```python
with ThreadPoolExecutor(max_threads) as executor:
    future_to_url = {executor.submit(get_image_src, url): url for url in filtered['url']}
    for future in as_completed(future_to_url):
        result = future.result()
        results.append(result)
```

```python
results
```

```python
dict_type = [item for item in results if isinstance(item, dict)]
str_type = [item for item in results if isinstance(item, str)]
str_type
```

```python
len(dict_type)
```

```python
list_url = [
    "https://www.goodreads.com/book/show/11492957-nicola-and-the-viscount",
    "https://www.goodreads.com/book/show/10752631-wild-life-vol-14",
    "https://www.goodreads.com//book/show/10787156-comanche-heart---akhir-penantian",
    "https://www.goodreads.com/book/show/9406111-grey-sunflower",
    "https://www.goodreads.com/book/show/10174603-kembalinya-sang-raja",
    "https://www.goodreads.com/book/show/9545606-nefertiti",
    "https://www.goodreads.com/book/show/7417859-kiss-me-annabel---cium-aku-annabel",
    "https://www.goodreads.com/book/show/6607024-sora-the-wingless-duck-vol-7",
    "https://www.goodreads.com/book/show/6606811-harry-potter-dan-pangeran-berdarah-campuran",
    "https://www.goodreads.com/book/show/2606518-rage-of-passion---hasrat-dan-cinta"
]
```

```python
resultan = []

for a in list_url:
    dicty = get_image_src(a)
    resultan.append(dicty)
```

```python
resultss = dict_type + resultan
```

```python
len(resultss)
```

```python
results_data = pd.DataFrame(resultss)
```

```python
len(results_data)
```

```python
book_fixing_image = filtered.merge(results_data, on='url', how='left')
book_fixing_image.info()
book_fixing_image.to_csv('Books_ID_Data_Fix.csv', index=False)
```
